```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#sns.get_dataset_names() 
```

```python
data = sns.load_dataset('taxis')
data.head()
```

```python
data.info()
```

```python
# 1, 2, 3 Plotting a histogram, dorpping na values, and changing the color 

sns.histplot(data = data , 
            x = data["total"].dropna(), # drop NA's on the series to plot
            color = "#3cb371") # change color of histogram

plt.show()
```

```python
# The plot that would allow you to plot two continuous features would be a line plot

sns.scatterplot(data = data, x = "distance", y = "total", hue = "color")
plt.show()
```

```python
# 5. 
corr = data[["total","tip", "distance"]].corr()
sns.heatmap(corr, annot=True, cmap='Blues')

plt.show()
```

```python
# 6. Change figure size using a matplotlib method figure

plt.figure(figsize=(10,8))
sns.scatterplot(data = data, x = "distance", y = "total", hue = "color")
plt.show()
```
